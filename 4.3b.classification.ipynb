```python
from transformers import pipeline

# zero-shot pipeline (юзає мульти-модель)
classifier = pipeline(
    "zero-shot-classification",
    model="joeddav/xlm-roberta-large-xnli"  # мульти-лінгвальний класифікатор
)

text = "Це чудовий день для вивчення нових технологій!"
labels = ["позитивний", "негативний", "нейтральний"]

result = classifier(text, candidate_labels=labels)
print(result)
print(f"Передбачений клас: {result['labels'][0]}")
```

```python
from transformers import pipeline

summarizer = pipeline(
    "summarization",
    model="ukr-models/uk-summarizer"
)

text = """
Українська мова посіла перше місце за темпами зростання серед відкритих моделей штучного інтелекту на Hugging Face у 2025 році.
Це свідчить про збільшення інтересу до інструментів обробки української мови, включно з перекладом, класифікацією та генерацією текстів.
"""

summary = summarizer(text, max_length=50, min_length=20)
print(summary)
```
